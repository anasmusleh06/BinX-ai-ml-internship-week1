```python
import pandas as pd
```

## Exercise 1

```python
df = pd.read_csv("../data/adult.csv")   # New info. for me: "..\" -> Means "go back folder".

print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\ndtypes:\n", df.dtypes)
```

## Exercise 2

```python
import numpy as np
df.replace("?", np.nan, inplace=True)  # New info. for me: replacing "?" with "NaN" to detect missing values.
Missing_counts = df.isnull().sum()     # Count missing values per column.

print("Missing Values per column:\n", Missing_counts[Missing_counts > 0])
```

## Exercise 3

```python
# The filter: individuals older than 50
above_50 = df[df["age"] > 50]

print("Number of individuals older than 50:", above_50.shape[0])
above_50.head()
```

## Exercise 4

```python
# Average hours worked per week are grouped by education level.
Avg_hours = df.groupby("education")["hours.per.week"].mean().sort_values(ascending=False)

print("Avarege Hours By Education:\n", Avg_hours)
```
